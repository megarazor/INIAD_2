```python
import matplotlib.pyplot as plt
#import matplotlib.pyplot.xticks as xtk


y= [12, 9, 8, 16, 20, 22, 20, 14]
x= ["2 PM", "5 PM", "8 PM", "11 PM", "14 PM", "17 PM", "20 PM", "23 PM"]
time_interval= range(0, len(x))


plt.plot(x, y, '-o')
plt.xticks(time_interval, x)
plt.axis(xmin=0, ymin=8)


plt.show()
```
