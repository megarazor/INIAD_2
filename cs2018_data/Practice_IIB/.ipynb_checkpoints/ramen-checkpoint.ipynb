```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('ramen-ratings.csv')
```

```python
df.shape
```

```python
df.info()
```

1. *Review* : unique numbers that inform the review order from the latest
2. *Brand*  : ramen brand
3. *Variety*: variation of ramen
4. *Style*  : style of ramen
5. *Country*: where the ramen is available
6. *Stars*  : ramen ratings
7. *Top Ten*: ramen achievement

```python
df.drop('Top Ten',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Style'])
```

```python
df.info()
```

```python
df = df.dropna(subset=['Stars'])
```

```python
df.info()
```

```python
df['Brand'].value_counts()
```

```python
sns.set(style = 'darkgrid')
f, ax = plt.subplots(1,1,figsize = (15,5))
sns.countplot(x = 'Brand', 
              data = df,  
              order = df['Brand'].value_counts()[:20].index)
plt.xticks(rotation=90)
plt.show()
```

```python
df['Variety'].value_counts()
```

```python
sns.set(style = 'darkgrid')
f, ax = plt.subplots(1,1,figsize = (15,5))
sns.countplot(x = 'Variety', 
              data = df,  
              order = df['Variety'].value_counts()[:20].index)
plt.xticks(rotation=90)
plt.show()
```

```python
df['Style'].value_counts()
```

```python
sns.set(style = 'darkgrid')
f, ax = plt.subplots(1,1,figsize = (15,5))
sns.countplot(x = 'Style', 
              data = df,  
              order = df['Style'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
df['Country'].value_counts()
```

```python
sns.set(style = 'darkgrid')
f, ax = plt.subplots(1,1,figsize = (15,5))
sns.countplot(x = 'Country', 
              data = df,  
              order = df['Country'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
df = df[df.Stars!='Unrated']
```

```python
df['Stars'] = df['Stars'].astype(float).sort_values(ascending=False)
```

```python
df[['Brand', 'Stars']].groupby(['Brand'], as_index=False).mean().sort_values(by="Stars",ascending=False)
```

```python
sns.set(style = 'darkgrid')
f, ax = plt.subplots(1,1,figsize = (15,5))
sns.barplot(x = 'Country', 
            y = 'Stars',
            data = df,  
            order = df['Country'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set(style = 'darkgrid')
f, ax = plt.subplots(1,1,figsize = (15,5))
sns.boxplot(x = 'Brand', 
              y = 'Stars',
              data = df,  
              order = df['Brand'].value_counts()[:20].index)
plt.xticks(rotation=90)
plt.show()
```
