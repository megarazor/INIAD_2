```python
import numpy as np
import matplotlib.pyplot as plt

x= np.linspace(1, 10, 100)

y1= x
y2= np.log(x)
#y3= x*np.log(x)

plt.plot(x, y1, color= 'red', label= '$y = x$')
plt.plot(x, y2, color= 'blue', label= '$y = \log x$')
#plt.plot(x, y3, color= 'green', label= '$y = x \log x$')

plt.xlim([1, 10])
plt.legend()
plt.xlabel('x', fontsize= 9)
plt.ylabel('y', fontsize= 9)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

x= np.linspace(1, 10, 100)

y1= x
y2= np.power(x, 2)
y3= x*np.log(x)

plt.plot(x, y1, color= 'red', label= '$y = x$')
plt.plot(x, y2, color= 'blue', label= '$y = x ^ 2$')
plt.plot(x, y3, color= 'green', label= '$y = x \log x$')

plt.xlim([1, 10])
plt.legend()
plt.xlabel('x', fontsize= 9)
plt.ylabel('y', fontsize= 9)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x= np.linspace(1, 10, 100)

y1= x
y2= np.power(x, 2)
y3= np.power(2, x)

plt.plot(x, y1, color= 'red', label= '$y = x$')
plt.plot(x, y2, color= 'blue', label= '$y = x ^ 2$')
plt.plot(x, y3, color= 'green', label= '$y = 2 ^ x$')

plt.xlim([1, 10])
plt.legend()
plt.xlabel('x', fontsize= 9)
plt.ylabel('y', fontsize= 9)
plt.show()
```

```python
import timeit
import matplotlib.pyplot as plt

def f(x):
    s= 0
    for i in range(x):
        for j in range(x):
            s+= 1
    return s

n= 1000

x= range(50, 500, 50)
y= []

for i in x:
    time= timeit.timeit('f(i)', globals= globals(), number= n)
    y.append(time/n)
    
plt.plot(x, y, 'o')
plt.xlabel('Number of iteration', fontsize= 9)
plt.ylabel('Average time (s)', fontsize= 9)

plt.show()
```

```python
import timeit
import matplotlib.pyplot as plt

def f(x):
    s= 0
    for i in range(x):
        for j in range(x):
            s+= 1
    for i in range(x):
        s+= 1
    return s

n= 1000

x= range(50, 500, 50)
y= []

for i in x:
    time= timeit.timeit('f(i)', globals= globals(), number= n)
    y.append(time/n)
    
plt.plot(x, y, 'o')
plt.xlabel('Number of iteration', fontsize= 9)
plt.ylabel('Average time (s)', fontsize= 9)

plt.show()
```
