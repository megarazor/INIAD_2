Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Read data

```python
elec=np.loadtxt('data/elec.csv', delimiter=',', skiprows=1)
```

Check data

```python
print( elec.shape )
print( elec )
```

Extract data

```python
mon=elec[:, 0]
e1=elec[:, 1]
e2=elec[:, 2]
```

Draw line plot

```python
plt.plot(mon, e1, label='2008')
plt.plot(mon, e2, label='2015')
plt.xlabel('Month')
plt.ylabel('Yen')
plt.legend()
plt.show()
```
