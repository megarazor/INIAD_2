Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

Reading Iris sepal data

```python
iris_sep=np.loadtxt('data/iris_sepal.csv', delimiter=',', skiprows=1)
```

Checking the data

```python
print( iris_sep.shape )
print( iris_sep[0:5] )
```

```python
x=iris_sep[:, 0]
y=iris_sep[:, 1]

print( x.shape )
print( x[0:5] )
print( y.shape )
print( y[0:5] )
```

```python
plt.scatter(x, y, marker='o', label='iris_sepal')
plt.xlabel('Length [cm]')
plt.ylabel('Width [cm]')
plt.legend()
plt.show()
```

```python
kmcls=KMeans(n_clusters=2).fit_predict(iris_sep)
print(kmcls)
```

```python
x0=x[kmcls==0]
y0=y[kmcls==0]
x1=x[kmcls==1]
y1=y[kmcls==1]
plt.scatter(x0, y0, marker='o', c='blue', label='cluster0')
plt.scatter(x1, y1, marker='D', c='red', label='cluster1')
plt.xlabel('Length [cm]')
plt.ylabel('Width [cm]')
plt.legend()
plt.show()
```
