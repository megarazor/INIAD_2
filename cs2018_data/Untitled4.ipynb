```python
import numpy as np
import matplotlib.pyplot as plt
import timeit
import random
```

```python
def setup_lst(x):
    lst = random.sample(range(x*2), k = x)
    c = random.randrange(x*2)
    return (lst,c)
def search_linear(lst, x):
    for i in range(len(lst)):
        if x == lst[i]: 
            return i
    return -1


def setup_lst_sorted(x):
    lst = sorted(random.sample(range(x*2),k=x))
    c = random.randrange(x*2)
    return (lst,c)
def search_bin(lst,x):
    lo = 1
    hi = len(lst)
    while lo < hi:
        mid = (lo+hi)//2
        if lst[mid-1]<x:
            lo = mid+1
        elif lst[mid-1]>x:
            hi = mid-1
        else:
            return mid
    return -1
```

```python
n = 1
r = 1000

xs = range(20,1000,20)
ys1 = []
ys2 = []

for x in xs:
    (lst,c) = setup_lst(x)
    time = timeit.repeat('search_linear(lst,c)',globals=globals(),number=n,repeat=r)
    ys1.append(sum(time)/(n*r))
    time = timeit.repeat('search_bin(lst,c)',globals=globals(),number=n,repeat=r)
    ys2.append(sum(time)/(n*r))

    
plt.plot(xs,ys1,color = 'red')
plt.ylim(0,max(ys1)*1.1)
plt.plot(xs,ys2,color = 'blue')
plt.xlabel('Length of list')
plt.ylabel('Average time(s)')
plt.show
```

```python
n = 1
r = 1000

xs = range(20,1000,20)
ys = []

for x in xs:
    time = timeit.repeat('search_bin(lst,c)',setup='(lst,c)=setup_lst_sorted(x)',globals=globals(),number=n,repeat=r)
    ys.append(sum(time)/(n*r))

    
plt.plot(xs,ys,color = 'red')
plt.ylim(0,max(ys)*1.1)
plt.xlabel('Length of list')
plt.ylabel('Average time(s)')
plt.show()
```
